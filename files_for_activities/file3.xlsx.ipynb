```python
#you can do replace for each of them, but there is an easier way by definition a function
f_vars = ["F","female","Female"]
m_vars = ["M", "male", "Male"]

def clear_gender_column(s):
    if s in f_vars:
        return "F"
    elif s in m_vars:
        return "M"
    else:
        return s
```

```python
#now we can apply the function to the column
df3["GENDER"] = df3["GENDER"].apply(clear_gender_column)
```

```python
#dropna drops "no values" or empty fields from a dataframe of a specified column
#when such empty fields are dropped, it removes the entire row of data where that empty cell exists. 
df3.dropna(subset=['GENDER'])
```

```python
#lambda functions (also called "annonymous" functions)
#you can use this inside of the "apply"
#it is good to use when you want to apply one function in a specific area and you don't need it anywhere else. 
sqr = lambda x: x*x
```
